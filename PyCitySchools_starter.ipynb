```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_number_schools = len(school_data_complete_df["school_name"].unique())
total_number_schools

total_number_students = (school_data_complete_df["Student ID"].count())
total_number_students

school_budget = (school_data_complete_df["budget"].unique())
total_budget = school_budget.sum()
total_budget

average_math_score = school_data_complete_df["math_score"].mean()
average_math_score

average_reading_score = school_data_complete_df["reading_score"].mean()  
average_reading_score

total_math_passing = school_data_complete_df.loc[school_data_complete_df["math_score"] >= 70, :]["math_score"].count()
total_math_passing
percent_math_passing = (total_math_passing/total_number_students)
percent_math_passing

total_reading_passing = school_data_complete_df.loc[school_data_complete_df["reading_score"] >= 70, :]["reading_score"].count()
total_reading_passing
percent_reading_passing = (total_reading_passing/total_number_students)
percent_reading_passing

total_math_reading_passing = school_data_complete_df.loc[(school_data_complete_df["math_score"] >= 70)&
                                                         (school_data_complete_df["reading_score"] >= 70), :]["Student ID"].count()
total_math_passing
percent_math_reading_passing =(total_math_reading_passing/total_number_students)
percent_math_reading_passing

District_Summary_DF = pd.DataFrame ({
    "Total Schools":[total_number_schools],
    "Total Students":[total_number_students],
    "Total Budget":[total_budget],
    "Average Math Score":[average_math_score],
    "Average Reading Score":[average_reading_score],
    "% Passing Math": [percent_math_passing],
    "% Passing Reading":[percent_reading_passing],
    "% Overall Passing":[percent_math_reading_passing]})

#https://stackoverflow.com/questions/63775903/python-jupyter-notebook-styled-dataframe-with-borders   

District_Summary_Formatted_DF=District_Summary_DF.style.format({'Total Budget': '${:,.2f}'.format,
                                                               'Total Students': '{:,.0f}'.format,
                                                                'Average Math Score': '{:,.6f}'.format,
                                                                'Average Reading Score': '{:,.5f}'.format,
                                                               '% Passing Math': '{:,.6%}'.format,
                                                               '% Passing Reading': '{:,.6%}'.format,
                                                               '% Overall Passing': '{:,.6%}'.format})

District_Summary_Formatted_DF
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
schools_grouped_df = school_data_complete_df.groupby(['school_name'])

gschool_type = school_data_complete_df.groupby(['school_name']).agg({"type":["max"]})
#Learned the droplevel concept from Tim Smith
gschool_type.columns = gschool_type.columns.droplevel(1)
#https://www.dataforeverybody.com/pandas-change-column-header-name/
gschool_type.rename(columns = {"type":"School Type"}, inplace="True")
gschool_type

gtotal_students = school_data_complete_df.groupby(['school_name']).agg({"student_name":["count"]})
gtotal_students.columns = gtotal_students.columns.droplevel(1)
gtotal_students.rename(columns = {"student_name":"Total Students"}, inplace="True")
gtotal_students

gtotal_budget = school_data_complete_df.groupby(['school_name']).agg({"budget":["max"]})
gtotal_budget.columns = gtotal_budget.columns.droplevel(1)
gtotal_budget.rename(columns = {"budget":"Total School Budget"}, inplace="True")
gtotal_budget

gper_student_budget = (gtotal_budget["Total School Budget"]/gtotal_students["Total Students"])   
gper_student_budget.columns =['Per Student Budget']
gper_student_budget

gaverage_math_score = school_data_complete_df.groupby(['school_name']).agg({"math_score":["mean"]})
gaverage_math_score.columns = gaverage_math_score.columns.droplevel(1)
gaverage_math_score.rename(columns = {"math_score":"Average Math Score"}, inplace="True")
gaverage_math_score

gaverage_reading_score = school_data_complete_df.groupby(['school_name']).agg({"reading_score":["mean"]})
gaverage_reading_score.columns = gaverage_reading_score.columns.droplevel(1)
gaverage_reading_score.rename(columns = {"reading_score":"Average Reading Score"}, inplace="True")
gaverage_reading_score

total_math_passing = school_data_complete_df.loc[school_data_complete_df["math_score"] >= 70, :]
total_math_passing_grouped = total_math_passing.groupby(['school_name']).agg({"student_name":["count"]})
total_math_passing_grouped
total_math_passing_grouped.columns = total_math_passing_grouped.columns.droplevel(1)
gpercent_math_passing= (total_math_passing_grouped["student_name"]/gtotal_students["Total Students"])
gpercent_math_passing

total_reading_passing = school_data_complete_df.loc[school_data_complete_df["reading_score"] >= 70, :]
total_reading_passing_grouped = total_reading_passing.groupby(['school_name']).agg({"student_name":["count"]})
total_reading_passing_grouped
total_reading_passing_grouped.columns = total_reading_passing_grouped.columns.droplevel(1)
gpercent_reading_passing= (total_reading_passing_grouped["student_name"]/gtotal_students["Total Students"])
gpercent_reading_passing

total_math_reading_passing = school_data_complete_df.loc[(school_data_complete_df["math_score"] >= 70)&
                                                         (school_data_complete_df["reading_score"] >= 70), :]
total_math_reading_passing_grouped = total_math_reading_passing.groupby(['school_name']).agg({"student_name":["count"]})
total_math_reading_passing_grouped
total_math_reading_passing_grouped.columns = total_math_reading_passing_grouped.columns.droplevel(1)
gpercent_math_reading_passing = (total_math_reading_passing_grouped["student_name"]/gtotal_students["Total Students"])
gpercent_math_reading_passing

# #https://stackoverflow.com/questions/40339886/pandas-concat-generates-nan-values                     
School_Summary = pd.concat([gschool_type,gtotal_students,gtotal_budget,gper_student_budget,
                            gaverage_math_score,gaverage_reading_score,gpercent_math_passing,
                            gpercent_reading_passing,gpercent_math_reading_passing
                           ], axis=1)

School_Summary.index.name=None

#https://re-thought.com/guide-to-renaming-columns-with-python-pandas/
School_Summary.rename(columns={0 : 'Per Student Budget', 1 : '% Passing Math', 2 : '% Passing Reading', 
                               3 : '% Overall Passing'}, inplace=True)

# School_Summary.reset_index()
School_Summary.columns


School_Summary_DF = School_Summary.style.format({'Total School Budget': '${:,.2f}'.format,
                                                 'Per Student Budget': '${:,.2f}'.format,
                                                'Total Students': '{:,.0f}'.format,
                                                 'Average Math Score': '{:,.6f}'.format,
                                                 'Average Reading Score': '{:,.6f}'.format,
                                                 '% Passing Math': '{:,.6%}'.format,
                                                 '% Passing Reading': '{:,.6%}'.format,
                                                 '% Overall Passing': '{:,.6%}'.format})

School_Summary_DF.columns
```
